### Задача 1.

В этой задаче нужно визуализировать *центральную предельную теорему*.

**a).** Пусть $\xi_1, ..., \xi_n$ --- независимые случайные величины из распределения $Exp(\lambda)$. Согласно центральной предельной теореме выполнена сходимость $$Z_n = \frac{X_n - \mathsf{E}X_n}{\sqrt{\mathsf{D}X_n}} \stackrel{d}{\longrightarrow} \mathcal{N}(0, 1),$$ где $X_n = \sum\limits_{i=1}^n \xi_i$. Вам нужно убедиться в этом, сгенерировав множество наборов случайных величин и посчитав по каждому из наборов величину $Z_n$ в зависимости от размера набора. 

Сгенерируйте 500 наборов случайных величин $\xi_1^j, ..., \xi_{300}^j$ из распределения $Exp(1)$. По каждому из них посчитайте сумму $X_{jn} = \sum\limits_{i=1}^n \xi_i^j$ для $1 \leqslant n \leqslant 300$, то есть сумма первых $n$ величин $j$-го набора. Для этого среднего посчитайте величину $$Z_{jn} = \frac{X_{jn} - \mathsf{E}X_{jn}}{\sqrt{\mathsf{D}X_{jn}}}.$$


```python
values = sps.expon(scale=1).rvs(size=[500, 300])
mean = sps.expon(scale=1).mean()
var = sps.expon(scale=1).var()
exp = (values.cumsum(axis=1) - np.full((500, 300), mean).cumsum(axis=1)) / np.sqrt(
    np.full((500, 300), var).cumsum(axis=1))
```

Для каждого $j$ нанесите на один график зависимость $Z_{jn}$ от $n$. Каждая кривая должна быть нарисована *одним цветом* с прозрачностью `alpha=0.05`. Сходятся ли значения $Z_{jn}$ к какой-либо константе?

```python
size = 300
count = 500
plt.figure(figsize=(15, 10))
for i in range(count):
    plt.plot(np.arange(size) + 1, exp[i], color='green', alpha=0.05)
plt.xlabel('Количество случайных величин')
plt.ylabel('Значение среднего')
plt.xlim((0, 300));
```

#### Нет, к какой-либо константе значения не сходятся

Для $n=300$ по набору случайных величин $Z_{1,300}, ..., Z_{500,300}$ постройте гистограмму. Похожа ли она на плотность распределения $\mathcal{N}(0, 1)$ (ее тоже постройте на том же графике)? Не забудьте сделать легенду.

```python
grid = np.linspace(-3, 3, 1000)

plt.figure(figsize=(15, 10))
plt.hist(exp[:, 299], label='Сгенерированные случайные величины', density=True)
plt.plot(grid, sps.norm.pdf(grid), lw=4, color='red', label='Плотность стандартного нормального распределения') 
plt.title('Сгенерированные случайные величины vs Cтандартное нормальное распределение')
plt.xlabel('Значение', fontsize=17)
plt.ylabel('Плотность', fontsize=17)
plt.legend()
plt.show()
```

**b).** Выполните те же действия для распределений $U(0, 1)$ и $Pois(1)$.

```python
values = sps.uniform().rvs(size=[500, 300])
mean = sps.uniform().mean()
var = sps.uniform().var()
uniform = (values.cumsum(axis=1) - np.full((500, 300), mean).cumsum(axis=1)) / np.sqrt(
    np.full((500, 300), var).cumsum(axis=1))
```

Сделайте вывод о смысле центральной предельной теоремы. Подтверждают ли сделанные эксперименты теоретические свойства?

### Задача 2.

В этой задаче нужно визуализировать *закон повторого логарифма*.

*a).* Пусть $\xi_1, ..., \xi_n$ --- независимые случайные величины из равномерного распределения на $\{-1, 1\}$. Согласно закону повторного логарифма траектория суммы $S_n = \xi_1 + ... + \xi_n$ при увеличении $n$ с вероятностью 1 бесконечное число раз пересекает границу $\pm (1-\varepsilon)\sqrt{2n\log\log n}, \varepsilon>0$, и лишь конечное число раз пересекает границу $\pm (1+\varepsilon)\sqrt{2n\log\log n}, \varepsilon>0$. Вам нужно убедиться в этом, сгенерировав множество наборов случайных величин и посчитав по каждому из наборов сумму в зависимости от размера набора.

Сгенерируйте 500 наборов случайных величин $\xi_1^j, ..., \xi_{300}^j$ из распределения $Bern(1/2)$. По каждому из них посчитайте среднее $S_{jn} = \sum\limits_{i=1}^n \xi_i^j$ для $1 \leqslant n \leqslant 300$, то есть сумму по первым $n$ величинам $j$-го набора.

```python
values = sps.bernoulli(p=0.5).rvs(size=(500, 300))
values[values == 0] = -1
S = values.cumsum(axis=1)
```

Для каждого $j$ нанесите на один график зависимость $S_{jn}$ от $n$. Каждая кривая должна быть нарисована *одним цветом* с прозрачностью `alpha=0.05`.

```python
size = 300
count = 500
plt.figure(figsize=(15, 10))
for i in range(count):
    plt.plot(np.arange(size) + 1, S[i], color='green', alpha=0.05)
plt.xlabel('Количество случайных величин')
plt.ylabel('Значение суммы')
plt.xlim((0, 300));
```

Для трех различных $n$ по набору случайных величин $S_{1,n}, ..., S_{500,n}$ постройте графих доли тех величин, которые выходят за порог $\pm с\sqrt{2n\log\log n}$ при изменении $c$ от 0.5 до 1.5. Графики стройте в строчку с помощью `plt.subplot`.

```python
import random
n = random.sample(range(1, 301), 3)
step = 1000
c = np.linspace(0.5, 1.5, step).reshape(step, 1)

plt.figure(figsize=(15, 15))

for i in range(3):
    tmp = S[:, n[i] - 1]
    tmp = np.tile(tmp, (step, 1))
    first_cond = tmp > c * np.sqrt(2 * n[i] * np.log(np.log(n[i])))
    second_cond = tmp < -c * np.sqrt(2 * n[i] * np.log(np.log(n[i])))
    tmp = (first_cond | second_cond).sum(axis=1) / 500
    
    plt.subplot(1, 3, i+1)
    plt.plot(c, tmp, lw=3)
    plt.xlabel('Значение с')
    plt.ylabel('Доля величин')
```

Сделайте вывод о смысле закона повторного логарифма. Подтверждают ли сделанные эксперименты теоретические свойства?

**Вывод:** закон повторного логарифма утверждает, что при случайном блуждании мы мочти наверное останемся в пределах $\pm с\sqrt{2n\log\log n}$ с увеличением n. Сделанные эксперименты **подтверждают** теоретические свойства.

**Будем исследовать рейтинг топ 250 фильмов по версии imdb и кинопоиска. Ожидаем, что большинство фильмов в рейтингах совпадут.** 

Начнем с imdb. Давайте сначала посмотрим на данные в чистом виде. Видно, что многие из них нуждаются в обработке. Давайте оставим только нужные для анализа колонки и приведем их к хорошему виду.

```python
imdb = pd.read_csv('imdb-top-250.csv')
imdb.head(5)
```

```python
necessary_columns = ['Rating', 'Original title', 'Release Date:', 'Director:', 'Stars:', 
                    'Genres:', 'Country:', 'Budget:', 'Cumulative Worldwide Gross:']
index = imdb[imdb['Original title'].isna()].index

for i in index:
    s = imdb['Title'][i]
    imdb['Original title'][i] = s[:len(s) - 8]
imdb = imdb[necessary_columns]
imdb.columns = ['Rating', 'Title', 'Release Year', 
                'Director', 'Stars', 'Genres', 'Country', 'Budget', 'Worldwide Gross']

for i in imdb['Rating'].index:
    rating = imdb['Rating'][i]
    imdb['Rating'][i] = rating[0:len(rating) - 3]

imdb['Rating'] = pd.to_numeric(imdb['Rating'])

for i in imdb['Title'].index:
    imdb['Title'][i] = imdb['Title'][i].replace(' (original title)', '')
    
for i in imdb['Release Year'].index:
    l = imdb['Release Year'][i].split(' ')
    for item in l:
        if len(item) == 4 and item.isdigit():
            imdb['Release Year'][i] = item
            break
            
imdb['Release Year'] = pd.to_numeric(imdb['Release Year']).astype(int)

for i in imdb['Stars'].index:
    imdb['Stars'][i] = imdb['Stars'][i].strip().replace(' See full cast & crew\xa0»', '')
```

```python
def clear_digit(str):
    l = list()
    for i in str:
        if i.isdigit():
            l.append(i)
    return ''.join(l)

def clear(column):
    for i in imdb[column].index:
        value = clear_digit(str(imdb[column][i]))
        if len(value) > 0:
            imdb[column][i] = int(value)
            
clear('Budget')
clear('Worldwide Gross')
```

Теперь поработаем с рейтингом кинопоиска. Снова отберем нужные стобцы и "почистим" данные.

```python
kinopoisk = pd.read_csv('kinopoisk-top-250.csv')
kinopoisk.head(5)
```

```python
necessary_columns = ['Рейтинг', 'Рейтинг кинокритиков', 'Оригинальное название', 'год', 'режиссер', 
                     'Главные роли', 'жанр', 'страна', 'бюджет', 'сборы в мире']

index = kinopoisk[kinopoisk['Оригинальное название'].isna()].index

for i in index:
    kinopoisk['Оригинальное название'][i] = kinopoisk['Название'][i]
    
kinopoisk = kinopoisk[necessary_columns]
kinopoisk.columns = ['Рейтинг', 'Рейтинг кинокритиков', 'Название', 'Год',
                'Режиссер', 'Главные роли', 'Жанр', 'Страна', 'Бюджет', 'Сборы в мире']
```

```python
for i in kinopoisk['Рейтинг'].index:
    kinopoisk['Рейтинг'][i] = np.round(kinopoisk['Рейтинг'][i], 1)
    
for i in kinopoisk['Рейтинг кинокритиков'].index:
    rating = str(kinopoisk['Рейтинг кинокритиков'][i])
    if rating != 'nan':
        kinopoisk['Рейтинг кинокритиков'][i] = float(rating[0:len(rating) - 1]) / 10

for i in kinopoisk['Главные роли'].index:
    kinopoisk['Главные роли'][i] = kinopoisk['Главные роли'][i].replace(', ..., ', '')
    
for i in kinopoisk['Жанр'].index:
    kinopoisk['Жанр'][i] = kinopoisk['Жанр'][i].replace(', ... слова', '')
def clear_digit(str):
    l = list()
    for i in str:
        if i.isdigit():
            l.append(i)
    return ''.join(l)

def clear(column):
    for i in kinopoisk[column].index:
        value = clear_digit(str(kinopoisk[column][i]))
        if len(value) > 0:
            kinopoisk[column][i] = int(value)
            
clear('Бюджет')

for i in kinopoisk['Сборы в мире'].index:
    value = str(kinopoisk['Сборы в мире'][i])
    l = list()
    if value[0] == '+':
        skip = True
        for item in value:
            if item == '=':
                skip = False
            if not skip and item.isdigit():
                l.append(item)
    else:
        for item in value:
            if item.isdigit():
                l.append(item)
    
    if ''.join(l) != '':
        kinopoisk['Сборы в мире'][i] = int(''.join(l))
```

### Задача 4.

Некоторые студенты второго курса ФИВТ понадеявшись на отмену учета посещения занятий по курсу "Введение в анализ данных" решили дудосить гугл-опросники. Команда "Физтех.Статистики" без особых проблем смогла разделить результаты опроса на спамовые и настоящие, а также установить круг подозреваемых. Теперь это предлагается сделать вам как начинающим аналитикам.

Вам выдаются результаты нескольких опросов. 
1. Необходимо для каждой строки понять, является ли результат спамовым или настоящим. Результаты анализа необходимо прислать на почту вместе с решением.
2. Какими общими характеристиками обладают спамовые записи? У многих спамовых записей отсутствует отчество, достаточно необычные имена. Как часто они происходят? 6-10 записей в секунду, в то время как настоящие записи происходят 2-3 раза в секунду (но если бы людей на курсе было бы больше, то частота могла бы быть одинаковой). Основная часть спамовых записей начинается с некоторого момента и идет последовательно (есть, конечно, люди, которые попали между спамом, но их мало).

```python
course = pd.read_csv('Список курса.csv')
course = course[['Unnamed: 1']].dropna().reset_index(drop=True).rename(columns={'Unnamed: 1': 'ФИ'})
course = course.dropna().drop(index=[0, 1])

for i in course['ФИ'].index:
    course['ФИ'][i] = ' '.join(str(course['ФИ'][i]).split(' ')[0 : 2])
    
course['ФИ'] = course['ФИ'].str.replace('ё', 'е')
course.reset_index(drop=True, inplace=True)
```

```python
april_30 = pd.read_csv('Лекция 30 апреля.csv')
april_30['ФИО'] = april_30['ФИО'].str.replace('ё', 'е').str.strip()
tmp_30 = april_30[['ФИО']].rename(columns={'ФИО': 'ФИ'})
    
for i in tmp_30['ФИ'].index:
    tmp_30['ФИ'][i] = ' '.join(str(tmp_30['ФИ'][i]).split(' ')[0 : 2])
    
visitors_30 = pd.merge(course, tmp_30, how='inner', left_on='ФИ', 
                    right_on='ФИ').drop_duplicates().reset_index(drop=True)

april_30['Спам'] = 'Да'
for i in april_30['ФИО'].index:
    l = ' '.join(str(april_30['ФИО'][i]).split(' ')[0 : 2])
    if l.strip() in np.array(visitors_30['ФИ']):
        april_30['Спам'][i] = 'Нет'
        
april_30.to_csv('Лекция 30 апреля (отфильтрованная).csv', encoding='utf-8')        
april_30
```

```python
march_5 = pd.read_csv('Лекция 5 марта.csv')
march_5['Фамилия Имя'] = march_5['Фамилия Имя'].str.replace('ё', 'е').str.strip()
tmp_5 = march_5[['Фамилия Имя']].rename(columns={'Фамилия Имя': 'ФИ'})
    
visitors_5 = pd.merge(course, tmp_5, how='inner', left_on='ФИ', 
                    right_on='ФИ').drop_duplicates().reset_index(drop=True)

march_5['Спам'] = 'Да'
for i in march_5['Фамилия Имя'].index:
    if march_5['Фамилия Имя'][i] in np.array(visitors_5['ФИ']):
        march_5['Спам'][i] = 'Нет'

march_5.to_csv('Лекция 5 марта (отфильтрованная).csv', encoding='utf-8')
march_5
```
